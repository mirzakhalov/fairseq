```python
import matplotlib.pyplot as plt
#from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, davies_bouldin_score,v_measure_score

import numpy as np
import pandas as pd
import os
import pickle
```

```python
def create_data():
    enc_df = pd.read_csv("enc_tok.csv", header=0,names = ['label', 'features'])
    features = enc_df['features'].values
    labels = enc_df['label'].values
    labels = np.array(labels)
    X = []

    for feature in features:
        temp = feature[1:-1].split()
        X.append([float(temp[0]), float(temp[1])])

    X = np.array(X)
    
    return labels, X

```

```python
def create_model(n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, init ='k-means++', max_iter=300, n_init=10,random_state=0)
    return kmeans
```

```python
def create_map(n_clusters, filename):
    labels, X = create_data()
    k_means = create_model(n_clusters)
    y_kmeans = k_means.fit_predict(X)
    language2cluster = dict()
    for i in range(0, len(labels)):
        language2cluster[labels[i]] = y_kmeans[i]
    
    with open(filename + '.pickle', 'wb') as handle:
        pickle.dump(language2cluster, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
    return language2cluster
```

```python
create_map(9, "map")
```
